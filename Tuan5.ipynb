```python
#W5A3
def snt(n):
    if n<2 :
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0 :
            return False
    return True
n = int(input())
print(snt(n))    
```

```python
#W5A4
def so_hoan_hao(n:int) -> bool:
    if n<2 :
        return False
    tong = 0
    for i in range(1,n):
        if n%i==0 :
            tong += i
    if tong == n :
        return True
    else:
        return False

assert so_hoan_hao(6) == True          
assert so_hoan_hao(9) == False
assert so_hoan_hao(28) == True
assert so_hoan_hao(496) == True
assert so_hoan_hao(10) == False

print("Passed all tests!")
```

```python
#W5A5
def timk(a,k):
    for i in range(len(a)):
        if a[i]==k :
            return i
    return -1
a = list(map(int,input().split()))
k = int(input())
print(timk(a,k))    
```

```python
#W5A6
def giai_thua(n):
    tich = 1
    for i in range(1,n+1):
        tich *= i
    return tich
n = int(input())
print(giai_thua(n))
```

```python
#W5A9
def tong_ccs(n:int) :
    tong = 0
    while n>0:
        tong += (n%10)
        n //= 10
    return tong
n = int(input())
print(tong_ccs(n))
```

```python
#W5A10
def is_isomorphic(a, b):
    map_a_b = {}
    map_b_a = {}
    
    for i in range(len(a)):
        x = a[i]
        y = b[i]
        
        if x in map_a_b:
            if map_a_b[x] != y:
                return False
        else:
            map_a_b[x] = y
        
        if y in map_b_a:
            if map_b_a[y] != x:
                return False
        else:
            map_b_a[y] = x
    
    return True
```

```python
#W5A12
def divide_with_remainder(dividend, divisor):
    q = dividend//divisor
    r = dividend%divisor
    return q,r
dividend,divisor = map(int,input().split())
print(divide_with_remainder(dividend, divisor))
```
